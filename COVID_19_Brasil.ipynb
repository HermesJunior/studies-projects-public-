# Panorama do COVID-19 no Brasil

O coronavírus pertence a uma família de vírus que causa desde resfriados comuns a doenças mais graves. O novo coronavírus recebeu a denominação SARS-CoV-2 pela Organização Mundial da Saúde (OMS). A doença que ele provoca tem a denominação COVID-19, que significa Coronavirus Disease (doença do coronavírus), enquanto “19” se refere ao ano em que apareceram os primeiros casos em Wuhan, na China.

Sabe-se que o vírus pode ser transmitido de pessoa a pessoa por meio do contato com secreções e posterior contato com a boca, nariz ou com os olhos. 

Os sintomas mais comuns são principalmente respiratórios sendo os sintomas mais comuns febre, tosse, dificuldades para respirar, cansaço, dor de garganta, dor de cabeça e náusea, vômito e diarréia.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos (30/06/2020).

*Fonte [Portal Covid-19](http://www.coronavirus.df.gov.br/)*

Até o presente momento, as informações são que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de administração de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar, exigindo equipamentos e profissionais treinados especificamente para isso.

O presente trabalho objetiva contribuir para elevar a consciência situacional a respeito do COVID-19 no Brasil. Os dados são de fontes oficiais e do *Our World in Data* publicadas na internet no final de junho/2020 (30/06/2020).

## Análise Exploratória dos Dados

Vamos importar os pacotes necessários e iniciar nossa Análise Exploratória.

```
# importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style()
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

# importar o dataset COVID-19 para um DataaFrame
#df = pd.read_csv('https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/owid-covid-data.csv')
df = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv')
```

Parte importante da nossa análise, é descobrir o tamanho do dataset que estamos encarando.

 Vamos verificar:

```
# verificando as dimensões
print('Dimensões do Dataset',
    '\nVariáveis: ',df.shape[1], "\n"
      'Entradas: ', df.shape[0])

# verificando as primeiras entradas
df.head()
```

```
# extrair os nomes das colunas
df.columns
```

```
# verificando o tipo das variáveis
df.dtypes
```

```
# verificando dados ausentes
round((df.isnull().sum() / df.shape[0]).sort_values(ascending=False),2)
```

Como vimos ao analisar o tipo das variáveis, verificamos que a coluna `date` não está em formato datetime. Para que possamos explorar essa variável ao máximo, vamos transformá-la para o formato datetime.

```
# transformando a coluna data em datetime
df.date = pd.to_datetime(df.date)
```

Após alterar o formato, vamos verificar qual a data mais recente do nosso dataset. Com essa informação, vamos verificar o país líder em casos naquele momento.

```
# verificando a data mais recente do dataset
df.date.max()
```

```
# total de casos
df.loc[df.date == '2020-06-30'].sort_values(by="total_cases", ascending=False)
```

Pudemos verificar que os países líderes em casos aos 30 de Junho de 2020 (19horas) eram:

* Estados Unidos - 2.094.069 casos
* Brasil - 1.368.195 casos
* Russia - 641156.0 casos

Agora, vamos verificar os casos de morte.

```
# total de mortes
df.loc[df.date == '2020-06-30'].sort_values(by="total_deaths", ascending=False)
```

Em relação às mortes, os países que lideravam aos 15 de Junho de 2020 eram:

* Estados Unidos - 126.140
* Brasil - 58.314
* Grã-Bretanha - 43.575 

Para melhorar a visualização, vamos plotar um gráfico de barras.

```
# gráfico de barras para total de mortes
top_5_total_deaths = df.loc[df.date == '2020-06-30', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[1:6]

fig, ax = plt.subplots(figsize=(15,10))

ax.bar('location', 'total_deaths', data=top_5_total_deaths)

plt.tight_layout()
```

### Como o vírus está progredindo?

Uma das maiores preocupações das organizações de saúde é a progressão do número de casos e mortes no mundo.

Por isso, vamos plotar um gráfico para demonstrar e evolução de mortes no mundo.

```
# gráfico de linha para mortes no mundo
df.loc[df.location == "World", "total_deaths"].reset_index(drop=True).plot(figsize=(9,7));
```

### Análise Exploratória para o Brasil

Agora que já temos uma noção dos acontecimentos no mundo, vamos focar nos acontecimentos no Brasil.

```
# Criar uma cópia do DataFrame apenas com o Brasil
df_brasil = df.loc[df.location == "Brazil"].copy()

df_brasil.head()
```

```
df_brasil.date.max()
```

Acima, verificamos que nosso dataset inclui datas desde **31-12-2019** até **30-06-2020**.

Vamos verificar quando foram identificados o primeiro caso e a primeira morte no Brasil.

```
# identificaçãso do primeiro caso
df_brasil.loc[df_brasil['new_cases'] != 0.0].head(1)
```

```
# Identificar quando foi registrada a primeira morte no Brasil
df_brasil.loc[df_brasil['new_deaths'] != 0.0].head(1) 
```

```
# Identificar quantos dias demorou para registrarmos a primeira morte, considerando o primeiro caso de COVID-19 registrado.
# identificaçãso do primeiro caso
print('O primeiro caso foi registrado em {}'.format(df_brasil.date.loc[3559]))
```

```
# primeira morte
print('A primeira morte foi registrada em {}'.format(df_brasil.date.loc[3580]))
```

Uma informação interessante para investigarmos é quanto tempo levou desde o primeiro caso até a primeira morte.

```
# Identificar quantos dias demorou para registrarmos a primeira morte, considerando o primeiro caso de COVID-19 registrado.
# calculando tempo desde o primeiro caso até a primeira morte
from datetime import date
delta = df_brasil.date.loc[3580] - df_brasil.date.loc[3559]
print('Se passaram {} dias entre o primeiro caso e a primeira morte no Brasil.'.format(delta.days))
```

Por fim, vamos plotar gráficos de linha para compararmos a evolução dos casos e mortes no Brasil.

```
# Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.
x = list(range(183))
z = df_brasil['total_cases']
z2 = df_brasil['total_deaths']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(9, 7))
ax1.set_title('Casos no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(x, z)

ax2.set_title('Casos no Brasil (Escala logarítmica)')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Casos')
ax2.set_yscale('log')
ax2.plot(x, z)

ax3.set_title('Mortes no Brasil')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Mortes')
ax3.plot(x, z2)

ax4.set_title('Mortes no Brasil (Escala logarítmica')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(x, z2)
plt.tight_layout()
```

```
# Graficos comparativos
# 
fig1, aux1 = plt.subplots(figsize=(9, 7))
df_brasil = df.loc[df.location == "Brazil"].copy()
df_usa = df.loc[df.location == "United States"].copy()
df_world = df.loc[df.location == "World"].copy()

xxx = list(range(183))
xx1 = df_usa['total_cases']
xx2 = df_brasil['total_cases']
xx3 = df_world['total_cases']

aux1.set_title('Total de Casos')
aux1.set_xlabel('Dias')
aux1.set_ylabel('Casos')
aux1.plot(xxx, xx1, label="USA")
aux1.plot(xxx, xx2, label="BRASIL")
aux1.plot(xxx, xx3, label="WORLD")
aux1.legend()

fig1.show()
```

```
# Graficos comparativos
# 
fig2, aux2 = plt.subplots(nrows=1, ncols=2, figsize=(20, 7))

xxx = list(range(183))
xx1 = df_usa['total_deaths']
xx2 = df_brasil['total_deaths']
xx3 = df_world['total_deaths']

# Grafico 1 total_deaths
aux2[0].set_title('Total de Mortes')
aux2[0].set_xlabel('Dias')
aux2[0].set_ylabel('Casos')
aux2[0].plot(xxx, xx1, label="USA")
aux2[0].plot(xxx, xx2, label="BRASIL")
aux2[0].plot(xxx, xx3, label="WORLD")
aux2[0].legend()



# Grafico 2 hospital_beds_per_thousand
xx4 = df_usa['hospital_beds_per_thousand']
xx5 = df_brasil['hospital_beds_per_thousand']
xx6 = df_world['hospital_beds_per_thousand']
aux2[1].set_title('Total de leitos Hospitalares (por mil)')
aux2[1].set_xlabel('Dias')
aux2[1].set_ylabel('Leitos/1000')
aux2[1].plot(xxx, xx4, label="USA")
aux2[1].plot(xxx, xx5, label="BRASIL")
aux2[1].plot(xxx, xx6, label="WORLD")
aux2[1].legend()

# Apresetnação Grafico1 + Grafico 2
fig2.show()
```
